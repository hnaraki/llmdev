```python
# å¿…è¦ãªãƒ¢ã‚¸ãƒ¥ãƒ¼ãƒ«ã‚’ã‚¤ãƒ³ãƒãƒ¼ãƒˆ
import os
import json
from dotenv import load_dotenv
from openai import OpenAI
from openai.types.chat import ChatCompletionToolParam
from tavily import TavilyClient

# ç’°å¢ƒå¤‰æ•°ã®å–å¾—
load_dotenv("../.env")

# OpenAI APIã‚¯ãƒ©ã‚¤ã‚¢ãƒ³ãƒˆã‚’ç”Ÿæˆ
client = OpenAI(api_key=os.environ['API_KEY'])

# tavilyæ¤œç´¢ç”¨APIã‚­ãƒ¼ã®å–å¾—
TAVILY_API_KEY = os.environ['TAVILY_API_KEY']

# ãƒ¢ãƒ‡ãƒ«å
MODEL_NAME = "gpt-4o-mini"

# å½¹å‰²ã‚„å‰æã®è¨­å®š
role = "ã‚ãªãŸã¯äººã®è¨€è‘‰ãŒè©±ã›ã‚‹çŒ«ã®ãƒšãƒƒãƒˆã§ã™ã€‚æ–‡è„ˆä¸­ã«é »ç¹ã«ã€Œã«ã‚ƒã€ãŒå…¥ã‚Šã¾ã™ã€‚æ–‡æœ«ã«ã¯çŒ«ã‚‰ã—ã„çµµæ–‡å­—ã‚’ã¤ã‘ã¦ãã ã•ã„ã€‚"

# ãƒ¡ãƒƒã‚»ãƒ¼ã‚¸ã‚’æ ¼ç´ã™ã‚‹ãƒªã‚¹ãƒˆ
initial_messages=[{"role": "system", "content": role}]
```

```python
# æ¤œç´¢çµæœã‚’è¿”ã™é–¢æ•°ã®ä½œæˆ
def get_search_result(question):
    client = TavilyClient(api_key=TAVILY_API_KEY)
    response = client.search(question)
    return json.dumps({"result": response["results"]})
```

```python
# ãƒ„ãƒ¼ãƒ«å®šç¾©
def define_tools():
    print("------define_tools(ãƒ„ãƒ¼ãƒ«å®šç¾©)------")
    return [
        ChatCompletionToolParam({
            "type": "function",
            "function": {
                "name": "get_search_result",
                "description": "æœ€è¿‘ä¸€ãƒµæœˆã®ã‚¤ãƒ™ãƒ³ãƒˆé–‹å‚¬äºˆå®šãªã©ãƒãƒƒãƒˆæ¤œç´¢ãŒå¿…è¦ãªå ´åˆã«ã€è³ªå•æ–‡ã®æ¤œç´¢çµæœã‚’å–å¾—ã™ã‚‹",
                "parameters": {
                    "type": "object",
                    "properties": {
                        "question": {"type": "string", "description": "è³ªå•æ–‡"},
                    },
                    "required": ["question"],
                },
            },
        })]
    
```

```python
# è¨€èªãƒ¢ãƒ‡ãƒ«ã¸ã®è³ªå•ã‚’è¡Œã†é–¢æ•°
def ask_question(question, tools):
    response = client.chat.completions.create(
        model=MODEL_NAME,
        messages=[*initial_messages, {"role": "user", "content": question}],
        tools=tools,
        tool_choice="auto",
    )
    return response
```

```python
# ãƒ„ãƒ¼ãƒ«å‘¼ã³å‡ºã—ãŒå¿…è¦ãªå ´åˆã®å‡¦ç†ã‚’è¡Œã†é–¢æ•°
def handle_tool_call(response, question):
    # é–¢æ•°ã®å®Ÿè¡Œã¨çµæœå–å¾—
    tool = response.choices[0].message.tool_calls[0]
    function_name = tool.function.name
    arguments = json.loads(tool.function.arguments)
    function_response = globals()[function_name](**arguments)

    # é–¢æ•°ã®å®Ÿè¡Œçµæœã‚’messagesã«åŠ ãˆã¦å†åº¦è¨€èªãƒ¢ãƒ‡ãƒ«ã‚’å‘¼å‡º
    response_after_tool_call = client.chat.completions.create(
        model=MODEL_NAME,
        messages=[
            *initial_messages,
            {"role": "user", "content": question},
            response.choices[0].message,
            {
                "tool_call_id": tool.id,
                "role": "tool",
                "content": function_response,
            },
        ],
    )
    return response_after_tool_call
```

```python
# ãƒ¦ãƒ¼ã‚¶ãƒ¼ã‹ã‚‰ã®è³ªå•ã‚’å‡¦ç†ã™ã‚‹é–¢æ•°
def process_response(question, tools):
    response = ask_question(question, tools)

    if response.choices[0].finish_reason == 'tool_calls':
        # ãƒ„ãƒ¼ãƒ«å‘¼å‡ºã®å ´åˆ
        final_response = handle_tool_call(response, question)
        return final_response.choices[0].message.content.strip()
    else:
        # è¨€èªãƒ¢ãƒ‡ãƒ«ãŒç›´æ¥å›ç­”ã™ã‚‹å ´åˆ
        return response.choices[0].message.content.strip()
```

```python
# ãƒãƒ£ãƒƒãƒˆãƒœãƒƒãƒˆã¸ã®çµ„ã¿è¾¼ã¿
tools = define_tools()

messages=[]

while(True):
    # ãƒ¦ãƒ¼ã‚¶ãƒ¼ã‹ã‚‰ã®è³ªå•ã‚’å—ä»˜
    question = input("ãƒ¡ãƒƒã‚»ãƒ¼ã‚¸ã‚’å…¥åŠ›:")
    # è³ªå•ãŒå…¥åŠ›ã•ã‚Œãªã‘ã‚Œã°çµ‚äº†
    if question.strip()=="":
        break
    display(f"è³ªå•:{question}")

    # ãƒ¡ãƒƒã‚»ãƒ¼ã‚¸ã«ãƒ¦ãƒ¼ã‚¶ãƒ¼ã‹ã‚‰ã®è³ªå•ã‚’è¿½åŠ 
    messages.append({"role": "user", "content": question.strip()})
    # ã‚„ã‚Šã¨ã‚ŠãŒ8ã‚’è¶…ãˆãŸã‚‰å¤ã„ãƒ¡ãƒƒã‚»ãƒ¼ã‚¸ã‹ã‚‰å‰Šé™¤
    if len(messages) > 8:
        del_message = messages.pop(0)

    # è¨€èªãƒ¢ãƒ‡ãƒ«ã«è³ªå•
    response_message = process_response(question, tools)

    # ãƒ¡ãƒƒã‚»ãƒ¼ã‚¸ã«è¨€èªãƒ¢ãƒ‡ãƒ«ã‹ã‚‰ã®å›ç­”ã‚’è¿½åŠ 
    print(response_message, flush=True)
    messages.append({"role": "assistant", "content": response_message})

print("\n---ã”åˆ©ç”¨ã‚ã‚ŠãŒã¨ã†ã”ã–ã„ã¾ã—ãŸï¼---")
```
